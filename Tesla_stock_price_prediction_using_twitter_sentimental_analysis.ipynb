###### ANN

```python
import tensorflow as tf       #importing libraries for deep learning model
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Dropout
```

```python
numpy_df_train.shape
```

```python
model=Sequential()                          #preparing custom model
model.add(Dense(10,activation='relu',input_dim=2))
model.add(Dropout(0.2))
model.add(Dense(40,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(160,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(250,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(100,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(12,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1,activation='relu'))
model.compile(optimizer='adam',loss='MeanAbsoluteError')
model.summary()
```

```python
history=model.fit(numpy_df_train, y_train,batch_size = 2, epochs= 100,shuffle=True)
```

```python
plt.plot(history.history['loss'])
#plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epochs')
plt.legend(['train', 'test'])
plt.show()
```
